### Graficar grupos de funciones

```python
x = np.linspace (-10, 10, 100)
y = (2*x) + 1
y2 = - x + 4
y3 = x**2
plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3)
```

```python
x = np.linspace (-10, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y)
plt.plot(x, y2)
```
